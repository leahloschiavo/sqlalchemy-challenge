```python
from sqlalchemy import create_engine, Table, Column, Integer, String, MetaData, inspect

engine = create_engine("sqlite:///C:/Users/leahm/Downloads/sqlalchemy-challenge/Starter_Code/Resources/hawaii.sqlite")

metadata = MetaData()

# Define a table
hawaii_measurements = Table('hawaii_measurements', metadata,
    Column('id', Integer, primary_key=True),
    Column('station', String),
    Column('date', String),
    Column('prcp', Integer),
    Column('tobs', Integer)
)

# Create the table in the database
metadata.create_all(engine)

# Use inspector to get the table names
inspector = inspect(engine)
print("Tables in the database:", inspector.get_table_names())
```

```python
# Save references to each table
from sqlalchemy import create_engine, Column, Integer, String, MetaData, ForeignKey
from sqlalchemy.ext.declarative import declarative_base

# Create the engine
engine = create_engine("sqlite:///C:/Users/leahm/Downloads/sqlalchemy-challenge/Starter_Code/Resources/hawaii.sqlite")

# Create a declarative base
Base = declarative_base()

# Define the hawaii_measurements table class
class HawaiiMeasurements(Base):
    __tablename__ = 'hawaii_measurements'
    
    id = Column(Integer, primary_key=True)
    station = Column(String)
    date = Column(String)
    prcp = Column(Integer)
    tobs = Column(Integer)

# Create the tables in the database
Base.metadata.create_all(engine)
```

```python
# Create our session (link) from Python to the DB
from sqlalchemy.orm import Session
session = Session(bind=engine)
#from class activity -- need to update
```

```python
# Inserting data
new_measurement = HawaiiMeasurements(station='XYZ', date='2023-11-14', prcp=1.5, tobs=75)
session.add(new_measurement)
session.commit()

# Querying data
result = session.query(HawaiiMeasurements).filter_by(station='XYZ').all()
for measurement in result:
    print(measurement.id, measurement.station, measurement.date, measurement.prcp, measurement.tobs)
```

# Exploratory Precipitation Analysis

```python
from sqlalchemy import desc

# Assuming you have already defined the HawaiiMeasurements class and created the session
from sqlalchemy import desc

# Query the data
result = session.query(HawaiiMeasurements).all()

# Find the most recent date
most_recent_date = max(measurement.date for measurement in result)
print(most_recent_date)
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results.
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import func
# Find the most recent date
most_recent_date = session.query(func.max(HawaiiMeasurements.date)).scalar()

# Starting from the most recent data point in the database.
most_recent_date = pd.to_datetime(most_recent_date)

# Calculate the date one year ago as a string
one_year_ago = (most_recent_date - pd.DateOffset(days=365)).strftime('%Y-%m-%d')

# Perform a query to retrieve the data and precipitation scores for the last 12 months
precipitation_data = session.query(HawaiiMeasurements.date, HawaiiMeasurements.prcp).\
    filter(HawaiiMeasurements.date >= one_year_ago).\
    order_by(HawaiiMeasurements.date).all()

# Save the query results as a Pandas DataFrame. Explicitly set the column names
df = pd.DataFrame(precipitation_data, columns=['Date', 'Precipitation'])

# Sort the DataFrame by date
df = df.sort_values(by='Date')

# Use Pandas Plotting with Matplotlib to plot the data
df.plot(x='Date', y='Precipitation', title='Precipitation Over the Last 12 Months')
plt.xlabel('Date')
plt.ylabel('Precipitation (inches)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
